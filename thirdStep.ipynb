```python
import pandas as pd
import os

def smellData(smellName):
    smellDF=pd.read_csv(os.path.join("ground truth",smellName+".csv"))
    return smellDF

def getClassDF(folder):
    classDF=pd.read_csv(os.path.join(folder,"allClass.csv"))
    classDF=classDF.drop(columns=["type_y"])
    classDF=classDF.rename(columns={"type_x":"type"})
    cols=list(classDF.columns)
    cols[10],cols[2]=cols[2],cols[10]
    classDF=classDF[cols]
    classDF["class"]=classDF["class"].str.replace("$",".")
    return classDF
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
pd.set_option("display.max_colwidth",None)
targetFolder="CoreNLP"      #
smellName="corenlp"

dataFolder=os.path.join(os.getcwd(),"data1")
folder=os.path.join(dataFolder,targetFolder)

classDF=getClassDF(folder)
```
